## reading html data

```python
import pandas as pd
import requests

url = "https://en.wikipedia.org/wiki/Economy_of_the_United_States"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
                  "(KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)
response.raise_for_status()   # ensures request was successful

df3 = pd.read_html(response.text)
df3
```

```python
len(df3)
```

```python
html_data = df3[3]
```

```python
html_data
```

```python
type(html_data)
```

```python
html_data.to_csv("US_GDP1.csv",index=False)
```
